# Importing Library

```python
import pandas as pd
import numpy as np

import scipy.stats as stats

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
```

#  Importing Dataset

```python
df = pd.read_csv('train.csv',usecols=['Age','Fare','Survived'])  # we need only 3 columns 
```

```python
df.head()
```

# Checking Null values And fill Them 

### checking Null Values

```python
sns.heatmap(df.isna())  # we have null values in age colums 
```

### Filling Null  Values 

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)  # filing age by mean of age 
```

```python
sns.heatmap(df.isna())   # no null values remaining 
```

# train_test_split

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop('Survived',axis=1),
                                                df['Survived'],test_size=0.2)
```

```python
X_train.head()
```

# Distribution And QQ Plot Of Age And Fare before Transformation 

### Distribution And QQ Plot Of Age

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
sns.distplot(X_train['Age'])
plt.title('Age PDF')

plt.subplot(122)
stats.probplot(X_train['Age'], dist="norm", plot=plt)
plt.title('Age QQ Plot')

plt.show()
```

The Age is Close to the normal Distribution in PDF graph 

The Blue line is just near to the red line means just near to normal distribution 

### Distribution And QQ Plot Of Fare

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
sns.distplot(X_train['Fare'])
plt.title('Fare PDF')

plt.subplot(122)
stats.probplot(X_train['Fare'], dist="norm", plot=plt)
plt.title('Fare QQ Plot')

plt.show()
```

The Fare pdf is Left Skewed and is not in normal form 

IN QQ plot The blue line is away from the red line so this is not the normal distribution 

# Model Training Before the Transformation 

```python
clf = LogisticRegression()   #Object for LogisticRegression
clf2 = DecisionTreeClassifier()  #Object for DecisionTreeClassifier
```

```python
clf.fit(X_train,y_train)  # apply  LogisticRegression on the data 
clf2.fit(X_train,y_train) # apply  DecisionTreeClassifier on the data 
    
y_pred = clf.predict(X_test)  # predict using LogisticRegression
y_pred1 = clf2.predict(X_test)  # predict using DecisionTreeClassifier
    
print("Accuracy LR",accuracy_score(y_test,y_pred))  # accuracy using LogisticRegression
print("Accuracy DT",accuracy_score(y_test,y_pred1))  # accuracy using DecisionTreeClassifier
```

accuracy using LogisticRegression is =  .6089385474860335

accuracy using DecisionTreeClassifier is  = 0.6480446927374302

# LOG TRANSFORMATION 

```python
trf = FunctionTransformer(func=np.log1p)  # making object for Functiontransformer in which applying function of log1p (log 1p means that log(value+1) because log 0 does not exist if in your data set there is not a 0 then you can use log only )
```

```python
X_train_transformed = trf.fit_transform(X_train)  # fitting this  log transformation on both X train y train data 
X_test_transformed = trf.transform(X_test)
```

# Model Training After the LOG Transformation 

```python
clf = LogisticRegression()
clf2 = DecisionTreeClassifier()

clf.fit(X_train_transformed,y_train)
clf2.fit(X_train_transformed,y_train)
    
y_pred = clf.predict(X_test_transformed)
y_pred1 = clf2.predict(X_test_transformed)
    
print("Accuracy LR",accuracy_score(y_test,y_pred))
print("Accuracy DT",accuracy_score(y_test,y_pred1))
```

The accuracy is better than the previous one because regression models gives more accuracy in Normal distribution

The accuracy of Decision tree does not depend on normal distribution so the accuracy is not much effected

# Reciprocal Transform  And Square Root Transform

### Function Of transformation 

```python
def apply_transform(transform):
    X = df.iloc[:,1:3]
    y = df.iloc[:,0]
    
    trf = ColumnTransformer([('log',FunctionTransformer(transform),['Fare'])],remainder='passthrough')
    
    X_trans = trf.fit_transform(X)
    
    clf = LogisticRegression()
    
    print("Accuracy",np.mean(cross_val_score(clf,X_trans,y,scoring='accuracy',cv=10)))
    
    plt.figure(figsize=(14,4))

    plt.subplot(121)
    stats.probplot(X['Fare'], dist="norm", plot=plt)
    plt.title('Fare Before Transform')

    plt.subplot(122)
    stats.probplot(X_trans[:,0], dist="norm", plot=plt)
    plt.title('Fare After Transform')

    plt.show()
    
```

### Reciprocal Transform

```python
apply_transform(lambda x : (1/x+0.002))  # for all the values in data frame reciprocal it 
```

### Square Root Transform

```python
apply_transform(lambda x :x**1/2 )   # for all the values oin data frame take square root of it 
```
